```julia
timing = DataFrame(Instance=String[], NNH_time=Float64[], MILP_time=Float64[], efficiency=Float64[])
# Before the loop
cost_summary = DataFrame(
    Instance        = String[],
    NNH_total_cost  = Float64[],
    MILP_total_cost = Float64[],
    Overall_gap     = Float64[],
    Mean_gap        = Float64[]
)
# Routing
## Initialization
for i in 1:7
    name = "p" * lpad(i, 2, "0")
    
    filepath = "MDVRP-Instances-master/dat/$name"
    inst = parse_mdvrp(filepath)
    instance_summary(inst)
    
    df = instance_to_dataframe(inst)
    
    static_df = create_static_df(inst)
    dynamic_df, depot_caps = create_dynamic_df(inst, static_df,df,name)
    
    grouping, vehicles_cap = create_grouping_df(inst, dynamic_df, name)
    
    ## NNH and MILP
    summary_df, routing_dict = prepare_routing_data(df, grouping, basename(filepath))
    
    #tsp = solve_tsp_nnh(routing_dict)
    #milp = solve_milp(routing_dict)

    nnh_time = @elapsed tsp = solve_tsp_nnh(routing_dict)
    milp_time= @elapsed milp = solve_milp(routing_dict)

    push!(timing, (
        Instance  = name,
        NNH_time  = round(nnh_time,  digits=4),
        MILP_time = round(milp_time, digits=4),
        efficiency  = round(milp_time / nnh_time)
    ))

    
    
    #comparison = compare_nnh_milp(tsp, milp)

    comparison, summary_row = compare_nnh_milp(tsp, milp, name)
    push!(cost_summary, summary_row)
    ## Routing Plots
     
    p_nnh = plot_groups(df, grouping, filepath)
    title!(p_nnh, basename(filepath) * " — NNH Routes")
    overlay_routes!(p_nnh, tsp, df)
    
    p_milp = plot_groups(df, grouping, filepath)
    title!(p_milp, basename(filepath) * " — MILP Routes")
    overlay_routes!(p_milp, rename(milp[:, [:vehicle_id, :milp_route]], :milp_route => :route), df)
    
    #display(plot(p_nnh, p_milp, layout=(1,2), size=(1500, 600)))

    combined = plot(p_nnh, p_milp, layout=(1,2), size=(1500, 600))
    display(combined)
    #savefig(combined, "Article/Figures/Routing/$(name)_routing.png")
    
    
end
```
